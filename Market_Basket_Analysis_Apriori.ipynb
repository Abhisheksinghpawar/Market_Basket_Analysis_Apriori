## MARKET BASKET ANALYSIS USING APRIORI

APRIORI is an example of unsupervised learnining algorithm used in association. Here i this project we will use it for market basket analysis for determining relation between the multiple items boght by a customer

```python
#Importing necessary libraries, Main library here is apyori library which will be used to do our market basket analysis

import pandas as pd
from apyori import apriori
```

```python
#Reading our data from csv file in to a dataframe using pandas library

data = pd.read_csv("D:\\Abhishek\\E Drive\\Padhai\\Programming\\Python\\Projects\\Market_Basket_Analysis_Apriori\\Market_Basket_Optimisation.csv")
```

This data has 7500 rows and 20 columns. we will be running 2 loops in order to get a list of lists of the items in our data. This list will be working as our training set from where we can generate our association rules.

```python
data_list = []
for i in range(1,len(data)):
    data_list.append([str(data.values[i,j]) for j in range(0,20)])
data_list[0:10]
```

Now that we have changed our dataset into a list of lists, we are ready to apply our apriori algorithm. We will define our minimum support, minimum confidence, minimum lift and minumum length

    - Support -> It is the frequecy of item A or a combination of (A and B)
    - Confindence -> It tells us that how often the combination of A and B occurs
    - Life -> It is the final strength of any rule
    - Length -> It is the min number of items there should be in any basket

```python
rules = apriori(data_list, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
results = list(rules)
```

This result file will contain all the sets and their respective confdence, support and lift values. We will print only the pairs in the next line of code

```python
for i in range(0, len(results)):
    print(results[i][0])
```

In the next line of code we will print each pair, it's support, confidence and life value respectively as per the index.

```python
for item in results:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    # second index of the inner list
    print("Support: " + str(item[1]))
    # third index of the list located at 0th position
    # of the third index of the inner list
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("-----------------------------------------------------")
```
